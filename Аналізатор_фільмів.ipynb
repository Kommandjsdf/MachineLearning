```python
def extract_genres(genres_str: str) -> list[str]:
  try:
    genres = ast.literal_eval(genres_str)
    return [genre["name"] for genre in genres]
  except ValueError:
    return []
```

```python
try:
  df['genres'] = df['genres'].apply(extract_genres)
except TypeError:
  df['genres'] = []
```

```python
genre_counts = df['genres'].value_counts() #Counts how many movies there are with every genre combination
print(genre_counts) #Prints it
```

```python
df.fillna({'budget': 0, 'revenue': 0}, inplace=True)

df['budget'] = pd.to_numeric(df['budget'], errors='coerce')

df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')

df['profit'] = df['revenue'] - df['budget']

df['profit']
```
