```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
np.random.seed(42)
X = np.vstack((np.random.randn(50, 2) + [2, 2], np.random.randn(50, 2) + [7, 7]))
```

```python
print(X)
```

```python
plt.scatter(X[:,0], X[:,1], s=50, color="gray")
# print(X[0,0], X[1,1])
# plt.scatter(X[0,0], X[0,1], s=50, color="blue")
plt.title("Output")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, random_state=24)
kmeans.fit(X)
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
print(labels)
print(centroids)
```

```python
plt.scatter(X[:,0],
            X[:,1],
            s=50,
            c=labels,
            cmap='viridis'
            )
plt.scatter(centroids[:,0],
            centroids[:,1],
            c="red",
            marker="X",
            s=200,
            label="Centroids"
            )
plt.title("Output")
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()
plt.show()
```
